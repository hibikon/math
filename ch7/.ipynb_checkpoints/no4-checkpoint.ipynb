◯　データから推測する

◯　移動平均を計算する

```python
# Numpyのconvolve()関数を使って移動平均を計算する

%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 気温データの読み込み
dat = pd.read_csv('temperature.csv', encoding="SHIFT-JIS")

# データ数
n = len(dat)
# x軸の値（１〜データ数）
x = range(1, n+1)

# 気温
# y軸の値（平均気温）
y = dat['平均気温']
# グラフを描画
plt.plot(x,y)

# 区間数：９の平均気温
v = np.ones(9) / 9.0
y2 = np.convolve(y, v, mode='same')
plt.plot(x[4:n-4], y2[4:n-4])
plt.show()
```

◯　回帰直線

◯　回帰直線の傾きと切片を求める

```python
# ２つのデータx[],y[]があるとき共分散は
import numpy as np

# xの平均値
mean_x = np.mean(x)
# yの平均値
mean_y = np.mean(y)
# xとyの共分散
cov = np.mean((x - mean_x) * (y - mean_y))
# で計算できる
```

```python
# xの分散は
var_x = np.var(x)
# で計算できる
```

```python
# これらのことから、回帰直線の傾きaと切片bは
a = cov / var_x
b = mean_y - (a * mean_x)
# で求められる
```

```python
# numpyのpolyfit()関数で回帰直線の計算ができる
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# データ
# 気温
x = ([23,24,28,24,27,21,18,25,28,20])
# ジュースの販売数
y = ([37,22,62,32,74,16,10,69,83,7])

# 回帰直線
a, b = np.polyfit(x, y, 1)
y2 = a * x + b

# 傾きと切片を表示
print('傾き：{0}, 切片：{1}'.format(a,b))

# 描画
# 散布図
plt.scatter(x,y)
# 回帰直線
plt.plot(x, y2)
plt.show()
```
