◯　微分

◯　変化を知る手がかり

◯　年収グラフと差分グラフ

```python
# 隣接する二つの年収の差分をとって、グラフを描く
# まずは年収グラフを描画

%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

# salary.csvの読み込み
dat = pd.read_csv('salary.csv', encoding='SHIFT-JIS')

# データをセット
x = dat['年齢']
y = dat['年収']

# グラフを描画
plt.plot(x, y)
plt.grid(color='0.8')
plt.show()
```

```python
# 今度は隣接する２つの年度の年収の差分をとり、その差分を使ってグラフを描画する

# データ数
cnt = len(dat)

# 差分をとる
diff_y = []
for i in range(0, cnt-1):
    diff_y.append(y[i+1] - y[i])
    
# グラフを描画
plt.plot(x[1:], diff_y)
plt.grid(color='0.8')
plt.show()
```
