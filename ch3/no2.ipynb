◯　方程式からグラフへ

◯　関数

```python
# 入力する値が決まると、出力値が１つに決まる式
# y = 3x -24 のxに１から１０までの値を入れていくと、yの値は次のように決まる

# 空のリストを作成
y = []

for x in range(1, 11):
    # y = 3x -24を計算して答えをリストに追加
    y.append(3 * x - 24)
    
y
```

```python
# グラフを描く
%matplotlib inline
import matplotlib.pyplot as plt

# データ
# xの値
x = list(range(1, 11))
y = []
for i in range(10):
    y.append(3 * x[i] -24)     # y = 3x -24
    
# グラフ
plt.plot(x, y)
plt.grid(color='0.8')
plt.show()
# y = 0のときが解
```

◯　式を使ってグラフを描画する

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# データ
x = np.arange(-1.0, 1.01, 0.01)
y = 3 * x

# グラフを描画
plt.plot(x, y)
plt.grid(color='0.8')
plt.show()
```
