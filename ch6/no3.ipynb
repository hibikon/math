◯　確率を計算する

```python
# サイコロをn回振った時に１が出る確率
# サイコロを振るかわりに乱数を使う
import random

# サイコロを振る
cnt = 0 # 1が出た回数
for i in range(10000):
    dice = random.randint(1,6)
    if dice == 1:
        cnt += 1
        
# 確率を求める
p = cnt / 10000
print(p)
```

◯　すべての目について確率を求める

```python
import random

# サイコロを振る
hist = [0] * 7
for i in range(10000):
    dice = random.randint(1,6)
    hist[dice] += 1
    
# 確率を求める
p = [0] * 7
for i in range(1, 7):
    p[i] = hist[i] / 10000
    print(i, p[i])

# 確率を合計
print('------------------\n' + str(sum(p)))
```

◯　積の法則と和の法則

```python
# ７本のうち２本当たりが入っているクジを考える
# 3番目にクジを引く人が当たる確率を求める
# 一番目に引く人の当たる確率は2/7, 3番目の人が当たる確率も2/7で同じになることを確かめる
# fractionsモジュールのFraction()クラスを利用すると分数の計算が可能

from fractions import Fraction

# 事象X（A当たり　→　Bはずれ　→　C当たり）
x = Fraction(2,7) * Fraction(5,6) * Fraction(1,5)
# 事象Y（Aはずれ　→　B当たり　→　C当たり）
y = Fraction(5,7) * Fraction(2,6) * Fraction(1,5)
# 事象Z（Aはずれ　→　Bはずれ　→　C当たり）
z = Fraction(5,7) * Fraction(4,6) * Fraction(2,5)
# X, Y, Zのいずれかが起こる確率
p = x + y + z
p
```

◯　モンテカルロ法で円周率を計算

```python
%matplotlib inline
import matplotlib.pyplot as plt
import random
import math

# 点を描画
cnt = 0
for i in range(3000):
    x = random.randint(1, 100)
    y = random.randint(1, 100)
    # 中心と点との距離
    d = math.sqrt((x-50)**2 + (y-50)**2)
    if (d <=50):
        # 円内の点を数える
        cnt += 1
        # 赤色の点を描画
        plt.scatter(x, y, marker='.', c='r')
    else:
        # 緑色の点を描画
        plt.scatter(x, y, marker='.', c='g')
plt.axis('equal')
plt.show()

# 円周率を求める
p = cnt / 3000
pi = p * 4
print(pi)
```
